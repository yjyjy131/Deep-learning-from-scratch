```python
class Score() :
  def __init__(self, mid, final) :
    self.__mid = mid
    self.__final = final

  @property 
  def mid(self) : return self.__mid

  @mid.setter
  def mid(self, mid) : self.__mid = mid


  @property 
  def final(self): return self.__final

  @final.setter
  def final(self, final) : self.__final = final



score = Score(50, 75)
print((score.mid + score.final)/2)
```

```python
class ReadCSV() :
  def __init__(self, file_path) :
    self.path = file_path
    self.lines = []


  def read_file(self) :
    f = open(self.path, 'r')
    lines = f.readlines()
    new_lines = []

    for line in lines: 
      new_lines.append(list(map(int, line[:-1].split(","))))

    f.close()
    self.lines = new_lines
    return new_lines


  def merge_list(self) :
    tmp_lines = []

    for line in self.lines :
      tmp_lines.append(sum(line))
      
    return tmp_lines


filepath = './data-01-test-score.csv'
read_csv = ReadCSV(filepath)

print(read_csv.read_file())
print(read_csv.merge_list())
```

```python
class ReadCSV() :
  def __init__(self, file_path) :
    self.path = file_path
    self.lines = []


  def read_file(self) :
    f = open(self.path, 'r')
    lines = f.readlines()
    new_lines = []

    for line in lines: 
      new_lines.append(list(map(int, line[:-1].split(","))))

    f.close()
    self.lines = new_lines
    return new_lines


  def merge_list(self) :
    f = open(self.path, 'r')
    lines = f.readlines()
    new_lines = []

    for line in lines: 
      new_lines.append(list(map(int, line[:-1].split(","))))

    tmp_lines = []
    all_len = len(new_lines[0])

    for line in new_lines :
      tmp_lines.append(sum(line)/all_len)
      
    tmp_lines.sort()
    return tmp_lines


filepath = './data-01-test-score.csv'
read_csv = ReadCSV(filepath)
print(read_csv.merge_list())
```
